```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('employment.txt', sep=",")
state_code = pd.read_csv('statecode.txt', sep='\t')
industry_code = pd.read_csv('Industrycode.txt', sep ='\t')
```

```python
print(data)
print(state_code)
print(industry_code)
```

```python
series_id = data['Series ID']
states_data = []
industry_data = []

state_data = []
industry_data = []

for index, row in data.iterrows():
    state_data.append(state_code[state_code['code'] == int(row['Series ID'][3:5])]['name'].iloc[0])
    industry_data.append(industry_code[industry_code['code'] == int(row['Series ID'][10:18])]['name'].iloc[0])
data['state'] = state_data
data['industry'] = industry_data
# print (industry_data)
```

```python
print (data)
```

```python
missing_var_in_state = {}
all_var = np.array(industry_code['name'])
for index, row in state_code['name'].iteritems():
    print (row)
    state_data = data[data['state'] == row]
    num_of_variables = len(state_data)
    print (num_of_variables)
    state_var = np.array(state_data['industry'])
    empty_var = [var for var in all_var if var not in state_var]
    missing_var_in_state[row] = {'number_of_missing': len(empty_var), 'missing_vars': empty_var}
    print ("Empty variables", empty_var)
    print ()

missing_state_in_var = {}
all_state = np.array(state_code['name'])
for index, row in industry_code['name'].iteritems():
    print (row)
    state_data = data[data['industry'] == row]
    num_of_states = len(state_data)
    print (num_of_states)
    var_state = np.array(state_data['state'])
    empty_state = [state for state in all_state if state not in var_state]
    missing_state_in_var[row] = {'number_of_missing': len(empty_state), 'missing_states': empty_state}
    print ("Empty state", empty_state)
    print ()
```

```python
print(missing_var_in_state)
print(missing_state_in_var)
```
